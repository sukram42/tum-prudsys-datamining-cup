```
import matplotlib.pyplot as plt
import numpy as np
```

```
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

```
plot_density = 50

price = 1
optimal_value = 50
margin = 3
```

```
def plot_data(x_l, x_u, ys):
  fig = plt.figure()  # figsize=(4,3))
  ax = fig.add_subplot(1, 1, 1)

  max_val = 0

  for y_l, y_u, c, label in ys:
    # plot the function
    plt.plot(x_u, y_u, c, label=label)
    plt.plot(x_l, y_l, c)

    max_y = max(y_u.max(), y_l.max())
    max_val = max(max_val, max_y)

  plt.axvline(x=optimal_value, linewidth=1, ls='--', color='black')

  text_x = optimal_value - 0.1
  text_y = max_val * 0.7
  plt.text(text_x, text_y, "optimum (example)", size=10, rotation=90.,
    ha="center", va="center")

  plt.ylabel('Loss')
  plt.xlabel('Predicted number of orders')

  plt.legend()
  plt.show()
```

```
lower_x = np.linspace(max(optimal_value-margin, 0), optimal_value, plot_density)
lower_y = (optimal_value - lower_x) * price
```

```
upper_x = np.linspace(optimal_value, optimal_value+margin, plot_density)
upper_y = (upper_x - optimal_value) * 0.6 * price
# plot_data(lower_x, upper_x, [(lower_y, upper_y, 'r')])
```

```
lower_y_squared = ((optimal_value - lower_x) * price)**2
upper_y_squared = ((upper_x - optimal_value) * 0.6 * price)**2
# plot_data(lower_x, upper_x, [(lower_y_squared, upper_y_squared, 'b')])
```

```
plot_data(lower_x, upper_x, [(lower_y, upper_y, 'r', 'Original loss'), (lower_y_squared, upper_y_squared, 'b', 'Squared loss')])
```

```
price_expensive = 2
lower_y_squared_expensive = ((optimal_value - lower_x) * price_expensive)**2
upper_y_squared_expensive = ((upper_x - optimal_value) * 0.6 * price_expensive)**2
```

```
plot_data(lower_x, upper_x, [(lower_y_squared_expensive, upper_y_squared_expensive, 'r', 'Price 2'), 
                             (lower_y_squared, upper_y_squared, 'b', 'Price 1')])
```
